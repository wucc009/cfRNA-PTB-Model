### 2.2 Calculation of Target Metrics

#### 2.2.1 Precision

```python
precision = TP / (TP + FP)
precision
```

#### 2.2.2 F1

```python
recall = TP / (TP + FN)
f1 = 2 * (precision * recall) / (precision + recall)
f1
```

#### 2.2.3 MCC

```python
import math
mcc = (TP * TN - FP * FN) / math.sqrt((TP + FP) * (TP + FN) * (TN + FP) * (TN + FN))
mcc
```
